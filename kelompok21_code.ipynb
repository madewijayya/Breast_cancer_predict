```python
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
import numpy as np 
```

import data training

```python
df_train = pd.read_csv('train.csv')
df_train = df_train.drop('id', axis=1)
df_train.head(10)
```

```python
df_train.info()
```

**PreProcessing**

```python
# melakukan checking terhadap missing value pada tiap kolom / features 
print('Jumlah Missing Values Pada Tiap features : \n')
print(df_train.isna().sum())
```

```python
#melihat kolom pada datset yang mengandung nilai ?
columns_with_question_mark = df_train.eq('?').any()
print(columns_with_question_mark)
```

```python
column = ['concavity_mean','concave points_mean','concavity_se','concave points_se','concavity_worst','concave points_worst'] # Daftar Kolom yang mengandung value '?'
for col in df_train.columns:
    count = (df_train[col] == '?').sum() 
    if count > 0:
        print(f"Kolom {col} dengan {count} nilai '?'")
for col in column: 
    mean_value = df_train[col].replace('?', np.nan).astype(float).mean()
    df_train[col] = df_train[col].replace('?', mean_value).astype(float)
```

```python
#memastikan apakah tidak ada nilai '?' yang tersisa 
for col in df_train.columns:
    count = (df_train[col] == '?').sum() 
    if count > 0:
        print(f"Kolom {col} memiliki {count} nilai '?'")
print('sudah tidak ada nilai "?" ')
```

```python
convert_diagnosis = {'M':1,'B':0}
df_train['diagnosis'] = df_train['diagnosis'].map(convert_diagnosis)
```

menampilkan boxplot agar dapat melihat pencilan pada data 

```python
def boxplot(df):
    sns.set(rc={'figure.figsize':(30,9)})
    sns.boxplot(x="variable", y="value", data=pd.melt(df[df.columns]))
    plt.xticks(rotation=90)
    plt.show()

boxplot(df_train)
```

```python
# Handling outlier with IQR method 
for i in df_train.columns:
  Q1 = df_train[i].quantile(0.25)
  Q3 = df_train[i].quantile(0.75)
  IQR = Q3-Q1

  Smallest = Q1 - (IQR *1.5) 
  Biggest = Q3 + (IQR * 1.5)

  df_train.loc[df_train[i] > Biggest , i] = Biggest 
  df_train.loc[df_train[i] < Smallest , i] = Smallest
```

hasil setelah dilakukan handling outlier 

```python
boxplot(df_train)
```

***import data testing*** 

**Pemodelan dan Eksperimen**

```python
df_test = pd.read_csv('test.csv')
df_test = df_test.drop('id', axis=1)
```

membagi data training dimana 
* x_train adalah kumpulan data yang akan menjadi input 
pada model nanti 

* y_train adalah berisi nilai target atau label untuk data pelatihan. 

x_train dan y_train akan berisi data pelatihan, sementara variabel x_val dan y_val akan berisi data validasi

20% dari data akan dialokasikan untuk data uji, sementara 80% akan digunakan sebagai data pelatihan

```python
# Spliting Data Training menjadi Data Validation
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(df_train.drop(['diagnosis'], axis=1), df_train['diagnosis'], test_size=0.2, random_state=42)
```

**Standarisasi**

untuk memastikan bahwa data uji memiliki skala yang sama dengan data pelatihan

```python
from sklearn.preprocessing import StandardScaler
# Scalling 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_val_scaled = scaler.transform(x_val)
```

**Instalasi dan penggunaan TPOT**

```python
#AutoML dengan TPOT 
!pip install tpot
from tpot import TPOTClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
tpot = TPOTClassifier(generations=10, cv=5, random_state=42, verbosity=2)
tpot.fit(x_train, y_train)
```

**Evaluasi**

```python
y_pred = tpot.predict(x_val)
accuracy = accuracy_score(y_val, y_pred)
# Evaluasi model dengan menghitung akurasi
print("Akurasi model pada data uji:", accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, y_pred))
```

```python
test_pred = tpot.predict(df_test)
df_test['diagnosis'] = test_pred
df_test.to_csv('result_prediction.csv', index=False)
```
